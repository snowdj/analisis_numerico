## Metodo de Newton para raices


```julia
using PyPlot

#defino la funci√≥n a utilizar

function f(x,a)
    return x.^2-a, 2*x;
end

```

```julia
a=2;
xx=(0:.01:2*a)

plot(xx,f(xx,a)[1])
grid()
```

```julia
x=4.0;
paso=Inf;

plot(xx,f(xx,a)[1])
grid()

while (abs(paso)/abs(x))>1e-10
    
    fx,dfx = f(x,a)
    plot(x,fx,"*")
    plot(xx,fx+(xx-x)*dfx,":")
    paso = -fx/dfx;
    x=x+paso;
    println(x)
end
plot(x,f(x,a)[1],"*");
```
