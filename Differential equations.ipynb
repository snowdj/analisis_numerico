# Ejemplos de ecuaciones diferenciales

## Se resuelve la ecuaci√≥n

$$ \dot{x} = f(t,x) $$


```julia
using PyPlot

f(x)=-11*x;

x0=2.0;
h=0.05;
Tspan = [0 1]
t=collect(Tspan[1]:h:Tspan[2]);
x=zeros(size(t))

x[1]=x0

for i=1:length(t)-1
    
    x[i+1]=x[i]+h*f(x[i])
    
end

plot(t,x,"*");
plot(t,x0*exp.(-11*t));
```

```julia
#metodo del trapecio

for i=1:length(t)-1

    #candidato
    aux=x[i]+h*f(x[i])
    err=Inf
    
    while(err>1e-8)
        aux=x[i]+h*(f(x[i])+f(aux))/2;
        err=x[i+1]-aux;
        x[i+1]=aux;
    end
    
end

plot(t,x,"*");
plot(t,x0*exp.(-11*t));
```
