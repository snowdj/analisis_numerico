```julia
using PyPlot

f(x) = exp.(x)-1;

h=logspace(-20,0,100);

df=f(h)./h;
E=abs.(df-1);


df2=(f(h)-f(-h))./(2*h);
E2=abs.(df2-1);

loglog(h,E,"*");
loglog(h,E2,"*");
```

```julia
g(x) = sin.(pi*x);

h=logspace(-20,0,100);

dg=g(h)./h;
Eg=abs.(dg-pi);


dg2=(g(h)-g(-h))./(2*h);
Eg2=abs.(dg2-pi);

loglog(h,Eg);
loglog(h,Eg2,"*");
```
